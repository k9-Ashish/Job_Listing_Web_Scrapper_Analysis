```python
# let's first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\Ashish\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url = "https://www.naukri.com"
driver.get(url)
```

```python
# finding element for job search bar
search_job = driver.find_element_by_xpath("//input[@class='sugInp']")
search_job
```

```python
# write on search bar
search_job.send_keys('Data Analyst')
```

```python
# finding element for job location bar
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore")
```

```python
search_btn= driver .find_element_by_xpath("//button[@class='btn']")
search_btn
```

```python
# clicking  using xpath function
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
#so let's extract all the tags having the job titles
title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
title_tags
```

```python
# extract the text of the job title  from the tags
job_titles=[]
for i in title_tags:
    if i.text is None:
        job_titles.append('Not')
    else:
        job_titles.append(i.text)
job_titles[:10]
```

```python
# lets extract all the tags having company names
company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company_tags
```

```python
# Now we will extract the text from the tags by looping over these tags
companies_names=[]

for i in company_tags:
    companies_names.append(i.text)
companies_names[:10]
```

```python
# so lets extract  all the tags having the experience required data
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience'] //span")
experience_tags
```

```python
# no we will extract the text  from these tags only by one by looping over these tags
experience_list=[]
for i in experience_tags:
    experience_list.append(i.text)
experience_list[:10]
```

```python
locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
locations_tags
```

```python
locations_list=[]
for i in locations_tags:
    locations_list.append(i.text)
locations_list[:10]
```

```python
#So lets check th length of ech element.
print(len(job_titles[:10])),print(len(companies_names[:10])),print(len(experience_list[:10])),print(len(locations_list[:10]))
```

# Creating a DataFarme for the Data Analyst jobs

```python
jobs=pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company']=companies_names[:10]
jobs['experience_required']=experience_list[:10]
jobs['location']=locations_list[:10]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df= pd.read_csv("/content/Naukri_Job_Data - Sheet1.csv")
```

```python
df
```

```python
df["title"].value_counts().plot(kind="bar")
```

```python
df["company"].value_counts().plot(kind="barh")
```

```python
# Company with the Most Job Listings (Bar chart)
top_company = df['company'].value_counts().idxmax()
company_counts = df['company'].value_counts()
company_counts.plot(kind='bar', figsize=(10, 6))
plt.title(f'Company with the Most Job Listings: {top_company}')
plt.xlabel('Company')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.show()
```

```python
# Company Distribution (Pie chart)
company_counts = df['company'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(company_counts, labels=company_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Job Listings by Company')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Bivariate Analysis: Scatter plot between experience_required and location count
location_counts = df['location'].value_counts()
plt.figure(figsize=(10, 6))
plt.scatter(df['experience_required'], df['location'].apply(lambda x: location_counts[x]), alpha=0.5)
plt.title('Bivariate Analysis: Experience vs. Location Count')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Jobs in Location')
plt.show()
```

```python
# Convert experience_required to numeric (taking the lower bound)
df['experience_required'] = df['experience_required'].str.extract('(\d+)').astype(float)

# Calculate the average years of experience required for each location
average_experience_by_location = df.groupby('location')['experience_required'].mean().reset_index()

# Create a bar chart to visualize the average experience required by location
plt.figure(figsize=(8, 4))
plt.bar(average_experience_by_location['location'], average_experience_by_location['experience_required'])
plt.title('Average Experience Required by Location')
plt.xlabel('Location')
plt.ylabel('Average Years of Experience Required')
plt.xticks(rotation=90)
plt.show()
```
